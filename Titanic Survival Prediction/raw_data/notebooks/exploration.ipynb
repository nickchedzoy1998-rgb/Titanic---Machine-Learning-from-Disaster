```python
def count_nans(df):
    na_df = df.isna().sum()
    return na_df

count_nans(train_df)
```

```python
count_nans(train_df)
```

```python
# Fill Missing Age Values by using most likley based on Class, Sex and Embarkment Location

train_df['Age'] = train_df.groupby(['Sex', 'Pclass', 'Embarked'])['Age'].transform(lambda x: x.fillna(x.median()))
```

```python
count_nans(train_df)
```

```python
# Two passengers cannot be filled as are missing both Age and Embarked
train_df[train_df['Age'].isna()]
```

```python
# Drop

train_df = train_df.dropna(subset=['Age', 'Embarked'])
count_nans(train_df)
```

```python
train_df['Cabin_cat'] = train_df['Cabin'].str[0]
train_df
```

```python
# Initial Stats

no_passengers = train_df['PassengerId'].nunique()
no_survivors = train_df['Survived'].sum()
perc_survived = (no_survivors / no_passengers) * 100
male_pas_no = len(train_df[train_df['Sex'] == 'male'])
fem_pas_no = len(train_df[train_df['Sex'] == 'female'])
male_survivors = len(train_df[(train_df['Sex'] == 'male') & (train_df['Survived'] == 1)])
fem_survivors = len(train_df[(train_df['Sex'] == 'female') & (train_df['Survived'] == 1)])
male_died = male_pas_no - male_survivors
fem_died = fem_pas_no - fem_survivors
male_surv_rate = (male_survivors / male_pas_no) *100
fem_surv_rate = (fem_survivors / fem_pas_no) * 100
class_1_tickets = len(train_df[train_df['Pclass'] == 1])
class_2_tickets = len(train_df[train_df['Pclass'] == 2])
class_3_tickets = len(train_df[train_df['Pclass'] == 3])
class_1_surv_rate = (len(train_df[(train_df['Pclass'] == 1) & (train_df['Survived'] == 1)]) / class_1_tickets) * 100
class_2_surv_rate = (len(train_df[(train_df['Pclass'] == 2) & (train_df['Survived'] == 1)]) / class_2_tickets) * 100
class_3_surv_rate = (len(train_df[(train_df['Pclass'] == 3) & (train_df['Survived'] == 1)]) / class_3_tickets) * 100



# Build a dictionary of your stats
stats = {
    "Metric": [
        "Total Passengers",
        "Total Survivors",
        "Percentage Survived",
        "Male Passengers",
        "Female Passengers",
        "Male Survivors",
        "Female Survivors",
        "Male Died",
        "Female Died",
        "Male Survival Rate (%)",
        "Female Survival Rate (%)",
        "First Class Tickets",
        "Second Class Tickets",
        "Third Class Tickets",
        "First Class Survivor Rate",
        "2nd Class Survivor Rate",
        "3rd Class Survivor Rate"
    ],
    "Value": [
        no_passengers,
        no_survivors,
        perc_survived,
        male_pas_no,
        fem_pas_no,
        male_survivors,
        fem_survivors,
        male_died,
        fem_died,
        male_surv_rate,
        fem_surv_rate,
        class_1_tickets,
        class_2_tickets,
        class_3_tickets,
        class_1_surv_rate,
        class_2_surv_rate,
        class_3_surv_rate
    ]
}

stats_df = pd.DataFrame(stats)

stats_df['Value'] = stats_df['Value'].round(1)

stats_df
```

```python
def age_bucket(age):
    if age <= 12:
        return 'Child'
    elif age <= 19:
        return 'Teen'
    elif age <= 39:
        return 'Adult'
    elif age <= 59:
        return 'Mid_Adult'
    else:
        return 'Senior'

train_df['age_bucket'] = train_df['Age'].apply(age_bucket)

age_buckets_survivor_rates = (
    train_df.groupby('age_bucket')['Survived']
    .mean() * 100
).round(1)

print(age_buckets_survivor_rates)

age_stats_df = pd.DataFrame({
    "Metric": [f"{bucket} Survival Rate" for bucket in age_buckets_survivor_rates.index],
    "Value": age_buckets_survivor_rates.values
})
```

```python
train_df['Cabin_cat'] = train_df.groupby(['Sex','Pclass','Embarked','age_bucket'])['Cabin_cat'] \
    .transform(lambda x: x.fillna(x.mode()[0] if not x.mode().empty else 'U'))
```

```python
count_nans(train_df)
```

```python
# Now usinf Cabin Cat instead of cabin so drop Cabin

train_df = train_df.drop(columns = 'Cabin')
```

```python
train_df['Fare'].describe()
```

```python
train_df['FareBucket'] = pd.qcut(train_df['Fare'], 4, labels=['Low','Mid','High','VeryHigh'])

fare_survival_rates = (
    train_df.groupby('FareBucket')['Survived']
    .mean() * 100
).round(1)

fare_stats_df = pd.DataFrame({
    "Metric": [f"{bucket} Fare Survival Rate" for bucket in fare_survival_rates.index],
    "Value": fare_survival_rates.values
})

```

```python
location_survival_rates = (
    train_df.groupby('Embarked')['Survived']
    .mean() * 100
).round(1)

embarkment_stats_df = pd.DataFrame({
    "Metric": [f"{location} Embarked Survival Rate" for location in location_survival_rates.index],
    "Value": location_survival_rates.values
})
```

```python
stats_df = pd.concat([stats_df, age_stats_df, fare_stats_df, embarkment_stats_df], ignore_index=True)
stats_df
```
